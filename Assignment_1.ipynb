```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
img_1 = io.imread("https://i.ibb.co/Kxbc9RQ/Whats-App-Image-2024-01-25-at-10-32-53-PM-1.jpg")
plt.imshow(img_1)
plt.title("Input Image 1")
plt.show()
imgGray_1 = cv2.cvtColor(img_1, cv2.COLOR_RGB2GRAY)
cv2_imshow(imgGray_1)

img_2 = io.imread("https://i.ibb.co/V2pqZSL/Whats-App-Image-2024-01-25-at-10-32-53-PM.jpg")
plt.imshow(img_2)
plt.title("Input Image 2")
plt.show()
imgGray_2 = cv2.cvtColor(img_2,cv2.COLOR_BGR2GRAY)
cv2_imshow(imgGray_2)
import urllib.request
from skimage import feature

def load_image(url):
    resp = urllib.request.urlopen(url)
    img = np.asarray(bytearray(resp.read()), dtype="uint8")
    img = cv2.imdecode(img, cv2.IMREAD_COLOR)
    return img

def apply_lbp(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    lbp = feature.local_binary_pattern(gray, P=8, R=1, method="uniform")
    return lbp

lbp_result_1 = apply_lbp(img_1)
lbp_result_2 = apply_lbp(img_2)

# Display the original given image 1
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img_1)
plt.title('Original Image')

# Display the LBP result image 1
plt.subplot(1, 2, 2)
plt.imshow(lbp_result_1, cmap='gray')
plt.title('LBP Result')

plt.show()


# Display the original given image 2
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img_2)
plt.title('Original Image')

# Display the LBP result image 2
plt.subplot(1, 2, 2)
plt.imshow(lbp_result_2, cmap='gray')
plt.title('LBP Result')

plt.show()
```
